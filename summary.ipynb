```python
# import modules 
import sys, os
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
df=pd.read_csv('conceptsdata.csv')
```

```python
df.columns
```

```python
df['NY']=df['States / U.S. Territories'].apply(lambda x: 1 if 'NY' in x else 0)
df['MA']=df['States / U.S. Territories'].apply(lambda x: 1 if 'MA' in x else 0)
df['VT']=df['States / U.S. Territories'].apply(lambda x: 1 if 'VT' in x else 0)
df['PA']=df['States / U.S. Territories'].apply(lambda x: 1 if 'PA' in x else 0)
df['NJ']=df['States / U.S. Territories'].apply(lambda x: 1 if 'NJ' in x else 0)
df['CT']=df['States / U.S. Territories'].apply(lambda x: 1 if 'NJ' in x else 0)
df['AI']=df['States / U.S. Territories'].apply(lambda x: 1 if 'AI' in x else 0)
df['reltext']=df['Submission Title']+df['Topic Summary']+df['Keywords']
df['reltext']=df['reltext'].astype(str)
df['reltext']=df['reltext'].apply(lambda x: x.lower())
df['artificial']=df['reltext'].apply(lambda x: x.count('artificial'))
df['intelligence']=df['reltext'].apply(lambda x: x.count('intelligence'))
df['climate']=df['reltext'].apply(lambda x: x.count('climate'))
df['health']=df['reltext'].apply(lambda x: x.count('health'))
df['ai']=df['reltext'].apply(lambda x: x.count(' ai '))
df['REG']=df[['NY','MA','VT','PA','NJ','CT']].sum(axis=1)
df['TOPIC']=df[['artificial','intelligence','climate','health','ai']].sum(axis=1)
df['POT_PARTNER']=df['REG']*df['TOPIC']

```

```python
df.loc[df['REG']>=1,:]
```

```python
pd.pivot_table(df, index=['Proposal Type'],values=['artificial','intelligence','climate','health','ai'],  aggfunc=['sum'])
```

```python
results_dict={}
results_dict['all']=df.sort_values(['NY','MA','VT','PA','NJ','CT'],ascending=False)
results_dict['locations']=pd.pivot_table(df, index=['Proposal Type'],values=['NY','MA','VT','PA','NJ','CT'],  aggfunc=['sum'])
results_dict['locations_select']=df.loc[df['REG']>=1,:].sort_values(['NY','MA','VT','PA','NJ','CT'], ascending=False)
results_dict['topics']=pd.pivot_table(df, index=['Proposal Type'],values=['artificial','intelligence','climate','health','ai'],  aggfunc=['sum'])
results_dict['topic_select']=df.loc[df['TOPIC']>=1,:].sort_values('TOPIC', ascending=False)
results_dict['partner']=df.loc[df['POT_PARTNER']>=1,:].sort_values('TOPIC', ascending=False)

```

```python
def write_excel(output, outfile):
    print("Outputing final Excel file:"+ str(outfile))
    with pd.ExcelWriter(outfile) as writer:
        for key, value in output.items():
            print("Outputting sheet:", key)
            value.to_excel(writer, sheet_name=key, index=True)


write_excel(results_dict, 'summary.xlsx')
    
    
```
